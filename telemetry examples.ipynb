```python
import fastf1
import matplotlib.pyplot as plt

plt.rcParams['axes.facecolor'] = '#262626'
plt.rcParams['figure.facecolor'] = 'black'
plt.rcParams['legend.facecolor'] = 'black'
plt.rcParams['legend.edgecolor'] = 'white'

plt.rcParams['text.color'] = 'white'
plt.rcParams['axes.labelcolor'] = 'white'
plt.rcParams['xtick.color'] = 'white'
plt.rcParams['ytick.color'] = 'white'
plt.rcParams['legend.facecolor'] = 'black'

ver_session = fastf1.get_session(2023, 'Monaco', 'Q')
ver_session.load()
ver_lap = ver_session.laps.pick_driver('VER').pick_fastest()
ver_tel = ver_lap.get_car_data().add_distance()

ham_session = fastf1.get_session(2023, 'Monaco', 'Q')
ham_session.load()
ham_lap = ham_session.laps.pick_driver('ALO').pick_fastest()
ham_tel = ham_lap.get_car_data().add_distance()

ver_distance = ver_tel['Distance']
ver_rpm = ver_tel['RPM']
ver_speed = ver_tel['Speed']
ver_throttle = ver_tel['Throttle']
ver_brake = ver_tel['Brake']
ver_n_gear = ver_tel['nGear']

ham_distance = ham_tel['Distance']
ham_rpm = ham_tel['RPM']
ham_speed = ham_tel['Speed']
ham_throttle = ham_tel['Throttle']
ham_brake = ham_tel['Brake']
ham_n_gear = ham_tel['nGear']

fig, axs = plt.subplots(5, 1, figsize=(13, 12), sharex=True, gridspec_kw={'hspace': 0.3})

axs[1].plot(ver_distance, ver_rpm, label='VER RPM', color='#b967ff')
axs[1].plot(ham_distance, ham_rpm, label='ALO RPM', color='#6FCB9F')
axs[1].set_ylabel('RPM')

axs[0].plot(ver_distance, ver_speed, label='Verstappen', color='#b967ff')
axs[0].plot(ham_distance, ham_speed, label='Alonso', color='#6FCB9F')
axs[0].set_ylabel('Speed (kmph)')
axs[0].legend(loc='upper right', bbox_to_anchor=(1.15, 1))

axs[2].plot(ver_distance, ver_throttle, label='VER Throttle', color='#b967ff')
axs[2].plot(ham_distance, ham_throttle, label='ALO Throttle', color='#6FCB9F')
axs[2].set_ylabel('Throttle')

axs[4].plot(ver_distance, ver_brake, label='VER Brake', color='#b967ff')
axs[4].plot(ham_distance, ham_brake, label='ALO Brake', color='#6FCB9F')
axs[4].set_ylabel('Brake')
axs[4].set_xlabel('Distance (meters)')

axs[3].plot(ver_distance, ver_n_gear, label='VER Gear', color='#b967ff')
axs[3].plot(ham_distance, ham_n_gear, label='ALO Gear', color='#6FCB9F')
axs[3].set_ylabel('Gear')

for ax in axs:
    ax.spines['bottom'].set_color('white')
    ax.spines['top'].set_color('white')
    ax.spines['right'].set_color('white')
    ax.spines['left'].set_color('white')

for ax in axs:
    ax.grid(color='white', linestyle='--', linewidth=0.5, alpha=0.7)

plt.tight_layout()

info_text = """
Max Verstappen = 1:11:365
Fernando Alonso = 1:11:449
"""
info_bbox = dict(boxstyle="round", edgecolor='#FFFF00', facecolor='black', alpha=1.0)
plt.text(0.5, 0.9, info_text, color='white', ha='center', va='bottom', bbox=info_bbox, fontsize=13, transform=fig.transFigure)

plt.show()
```
