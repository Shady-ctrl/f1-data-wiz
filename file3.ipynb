```python
from matplotlib import pyplot as plt
import fastf1
import fastf1.plotting

fastf1.plotting.setup_mpl()

session = fastf1.get_session(2023, 'singapore', 'Q')

session.load()
fast_leclerc = session.laps.pick_driver('VER').pick_fastest()
lec_car_data = fast_leclerc.get_car_data()
t = lec_car_data['Time']
vCar = lec_car_data['Speed']

fast_leclerc1 = session.laps.pick_driver('LEC').pick_fastest()
lec_car_data1 = fast_leclerc1.get_car_data()
t1 = lec_car_data1['Time']
vCar1 = lec_car_data1['Speed']

fig, ax = plt.subplots(figsize=(12, 8)) 

plt.plot(t, vCar, label='Verstappen', color='blue')
plt.plot(t1, vCar1, label='Leclerc', color='red')

fig.patch.set_facecolor('black')
ax.set_xlabel('Time (s)', fontsize=12)  
ax.set_ylabel('Speed (Km/h)', fontsize=12)  
ax.set_title('Head-to-Head Speed Analysis\nVerstappen vs. Leclerc', fontsize=14, color='white')  

ax.grid(linestyle=':', linewidth=0.5, alpha=0.3)

ax.legend(loc='upper right').get_texts()[0].set_color('white')
ax.legend(loc='upper right').get_texts()[1].set_color('white')

plt.show()


#for further analysis, visit:
#https://x.com/adarsh82_/status/1742187445092007954 
```
